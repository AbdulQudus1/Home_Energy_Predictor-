```python
import matplotlib.pyplot as plt

df['EnergyConsumption'].hist(bins=30, color='skyblue')
plt.title('Energy Consumption Distribution')
plt.xlabel('EnergyConsumption')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```

```python
# Create a mapping dictionary
day_mapping = {
    'Sunday': 0,
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6
    
}

# Apply the mapping to the DayOfWeek column
df['DayOfWeek'] = df['DayOfWeek'].map(day_mapping)

# Verify the result
df[['DayOfWeek']].head()
```

```python
# Mapping dictionaries for each column
holiday_map = {'No': 0, 'Yes': 1}
hvac_map = {'Off': 0, 'On': 1}
lighting_map = {'Off': 0, 'On': 1}

# Apply the mappings
df['Holiday'] = df['Holiday'].map(holiday_map)
df['HVACUsage'] = df['HVACUsage'].map(hvac_map)
df['LightingUsage'] = df['LightingUsage'].map(lighting_map)

# Check the transformed columns
df[['Holiday', 'HVACUsage', 'LightingUsage']].head()
```
